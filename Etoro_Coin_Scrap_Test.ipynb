```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
```

```python
# Configures webdriver
#
# Hides GUI can solve the problem of web page stucking on loadding

options = Options()
options.headless = True # Hides GUI

# Sets window size to native GUI size
# options.add_argument("--window-size=1920,1080") 
# options.add_argument("start-maximized")

# Removes automation tags from the browser
# options.add_experimental_option("excludeSwitches", ["enable-automation"])
# options.add_experimental_option('useAutomationExtension', False)

# Reference:
# https://scrapfly.io/blog/web-scraping-with-selenium-and-python/
# https://python.plainenglish.io/how-to-bypass-google-recaptcha-with-selenium-and-python-9470664fb575
```

```python
# Initializes driver

service = Service('C:/Users/ZH128895/chromedriver.exe') # Your/Absolute/Path/To/chromedriver.exe 
driver = webdriver.Chrome(service=service, options = options) # Uses Chrome brower

# Changes User Agent at each request, cannot execute without this line, not sure why 
driver.execute_cdp_cmd('Network.setUserAgentOverride', {"userAgent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.53 Safari/537.36'})

driver.get('https://www.etoro.com/discover/markets/cryptocurrencies/coins') # Website to scrap
```

```python
# Finds two buttons elements by XPATH

next_page_btn = driver.find_element(By.XPATH, '/html/body/ui-layout/div/et-layout-main/div/div[2]/div[2]/div[3]/div/ng-view/et-discovery-markets-results/div/et-discovery-markets-results-header/div/div[2]/div/div[2]/a[2]')

prev_page_btn = driver.find_element(By.XPATH, '/html/body/ui-layout/div/et-layout-main/div/div[2]/div[2]/div[3]/div/ng-view/et-discovery-markets-results/div/et-discovery-markets-results-header/div/div[2]/div/div[2]/a[1]')
```

```python
# Navigates to next page then return back
#
# The reason to perform this action is there is one coin item not loaded the first time loads to the site,
# the missing coin item is loaded after navigating to the next page then navigates back

# Creates action chain object
action = ActionChains(driver)

# Clicks the next page button
action.click(on_element = next_page_btn)

# Clicks the previous page button
action.click(on_element = prev_page_btn)

# Performs the operation
action.perform()
```

```python
# Extracts the coin info text


# Init two lists for storing coin symbol and coin name respectively
coin_symbol_lst = []
coin_name_lst = []

# Scraps the coin info from the first page
coin_symbols_1 = driver.find_elements(By.CSS_SELECTOR, 'div.symbol')
coin_names_1 = driver.find_elements(By.CSS_SELECTOR, 'div.name')

for index in range(len(coin_symbols_1)):
    coin_symbol_lst.append(coin_symbols_1[index].text)
    coin_name_lst.append(coin_names_1[index].text)
    
# Navigates to next page
action = ActionChains(driver)

action.click(on_element = next_page_btn)

action.perform()

# Scraps the coin info from the second page
coin_symbols_2 = driver.find_elements(By.CSS_SELECTOR, 'div.symbol')
coin_names_2 = driver.find_elements(By.CSS_SELECTOR, 'div.name')

for index in range(len(coin_symbols_2)):
    coin_symbol_lst.append(coin_symbols_2[index].text)
    coin_name_lst.append(coin_names_2[index].text)
```
